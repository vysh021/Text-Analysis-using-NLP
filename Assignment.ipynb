```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import nltk
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import string
from textblob import TextBlob
import csv
import numpy as np
```

```python
link_df=pd.read_excel("Input.xlsx")
link_df.head()
```

```python
articles=[]
for i in link_df['URL']:
    page=requests.get(i)
    soup = BeautifulSoup(page.text, 'html.parser')
    article=(soup.find("div",{"class":"td-post-content tagdiv-type"}).text).replace('\n','').replace('\xa0','')
    articles.append(article)
```

```python
import re
import string
def preprocess_text(text):
    # Lowercase the text
    text = text.lower()

    # Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)

    # Remove email addresses
    text = re.sub(r'\S+@\S+', '', text)

    # Remove special characters and punctuation
    text = re.sub(r'[%s]' % re.escape(string.punctuation), ' ', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()

    text = text.replace('‚óè', '')

    return text
```

```python
cleaned_articles=[]
for i in articles:
    cleaned_articles.append(preprocess_text(i))
```

```python
stop_words=[]
file= open('StopWords_Geographic.txt', 'r')
read=file.readlines()
for line in read:
  if line[-1]=='\n':
    stop_words.append(line[:-1])
  else:
    stop_words.append(line)
for item in stop_words:
  item.replace('\n ','')
file.close()
file= open('StopWords_Auditor.txt', 'r')
read=file.readlines()
for line in read:
  if line[-1]=='\n':
    stop_words.append(line[:-1])
  else:
    stop_words.append(line)
for item in stop_words:
  item.replace('\n ','')
file.close()
file= open('StopWords_DatesandNumbers.txt', 'r',encoding = "ISO-8859-1")
read=file.readlines()
for line in read:
  if line[-1]=='\n':
    stop_words.append(line[:-1])
  else:
    stop_words.append(line)
for item in stop_words:
  item.replace('\n ','')
file.close()
file= open('StopWords_Currencies.txt', 'r',encoding = "ISO-8859-1")
read=file.readlines()
for line in read:
  if line[-1]=='\n':
    stop_words.append(line[:-1])
  else:
    stop_words.append(line)
for item in stop_words:
  item.replace('\n ','')
file.close()
file= open('StopWords_Generic.txt', 'r',encoding = "ISO-8859-1")
read=file.readlines()
for line in read:
  if line[-1]=='\n':
    stop_words.append(line[:-1])
  else:
    stop_words.append(line)
for item in stop_words:
  item.replace('\n ','')
file.close()
file= open('StopWords_GenericLong.txt', 'r',encoding = "ISO-8859-1")
read=file.readlines()
for line in read:
  if line[-1]=='\n':
    stop_words.append(line[:-1])
  else:
    stop_words.append(line)
for item in stop_words:
  item.replace('\n ','')
file.close()
file= open('StopWords_Names.txt', 'r',encoding = "ISO-8859-1")
read=file.readlines()
for line in read:
  if line[-1]=='\n':
    stop_words.append(line[:-1])
  else:
    stop_words.append(line)
for item in stop_words:
  item.replace('\n ','')
file.close()
stop_words = [word.lower() for word in stop_words]
stop_words=set(stop_words)
```

```python
for i in range(len(cleaned_articles)):
  for w in stop_words:
        cleaned_articles[i]= articles[i].replace(' '+w+' ',' ')
```

```python
sentences = []
for article in articles:
  sentences.append(len(sent_tokenize(article)))
```

```python
words = []
for article in articles:
  words.append(len(word_tokenize(article)))
#print(words)
```

```python
words_cleaned = []
for article in cleaned_articles:
  words_cleaned.append(len(word_tokenize(article)))
```

```python
words_cleaned = np.array(words_cleaned)
sentences = np.array(sentences)
```

```python
Output_data_structure=pd.read_excel("Output Data Structure.xlsx")
Output_data_structure.head()
```

```python
file= open('positive-words.txt', 'r')
read=file.readlines()
positive_words=[]
for line in read:
  if line[-1]=='\n':
    positive_words.append(line[:-1])
  else:
    positive_words.append(line)
for item in positive_words:
  item.replace('\n ','')
file.close()
```

```python
file= open('negative-words.txt', 'r',encoding='latin-1')
read=file.readlines()
negative_words=[]
for line in read:
  if line[-1]=='\n':
    negative_words.append(line[:-1])
  else:
    negative_words.append(line)
for item in negative_words:
  item.replace('\n','')
file.close()
```

```python
positive_words = [word.lower() for word in positive_words]
```

```python
positive_score = [0]*len(articles)
for i in range(len(articles)):
  for word in positive_words:
    for letter in cleaned_articles[i].lower().split(' '):
      if letter==word:
        positive_score[i]+=1
```

```python
negative_score = [0]*len(articles)
for i in range(len(articles)):
  for word in negative_words:
    for letter in cleaned_articles[i].lower().split(' '):
      if letter==word:
        negative_score[i]+=1
```

```python
Output_data_structure['POSITIVE SCORE'] = positive_score
Output_data_structure['NEGATIVE SCORE'] = negative_score
```

```python
Output_data_structure['POLARITY SCORE'] = (Output_data_structure['POSITIVE SCORE']-Output_data_structure['NEGATIVE SCORE'])/ ((Output_data_structure['POSITIVE SCORE'] +Output_data_structure['NEGATIVE SCORE']) + 0.000001)
```

```python
Output_data_structure['SUBJECTIVITY SCORE'] = (Output_data_structure['POSITIVE SCORE'] + Output_data_structure['NEGATIVE SCORE'])/( (words_cleaned) + 0.000001)
```

```python
Output_data_structure['AVG SENTENCE LENGTH'] = np.array(words)/np.array(sentences)
```

```python
complex_words = []
sylabble_counts = []
```

```python
for article in articles:
  sylabble_count=0
  d=article.split()
  ans=0
  for word in d:
    count=0
    for i in range(len(word)):
      if(word[i]=='a' or word[i]=='e' or word[i] =='i' or word[i] == 'o' or word[i] == 'u'):
           count+=1
#            print(words[i])
      if(i==len(word)-2 and (word[i]=='e' and word[i+1]=='d')):
        count-=1;
      if(i==len(word)-2 and (word[i]=='e' and word[i]=='s')):
        count-=1;
    sylabble_count+=count
    if(count>2):
        ans+=1
  sylabble_counts.append(sylabble_count)
  complex_words.append(ans)
```

```python
Output_data_structure['PERCENTAGE OF COMPLEX WORDS'] = np.array(complex_words)/np.array(words)
```

```python
# Function to calculate average words per sentence
def avg_words_per_sentence(text):
    # Split the text into sentences
    sentences = re.split(r'[.!?]', text)
    # Remove empty strings from the list
    sentences = [s.strip() for s in sentences if s.strip()]
    # Calculate the number of words in each sentence
    num_words = [len(s.split()) for s in sentences]
    # Calculate the average number of words per sentence
    if len(num_words) > 0:
        return sum(num_words) / len(num_words)
    else:
        return 0
count=[]
for article in articles:
  count.append(avg_words_per_sentence(article))
Output_data_structure['AVG NUMBER OF WORDS PER SENTENCE'] =count
```

```python
total_characters = []
for article in articles:
  characters = 0
  for word in article.split():
    characters+=len(word)
  total_characters.append(characters)
#print(total_characters)
```

```python
personal_nouns = []
personal_noun =['I', 'we','my', 'ours','and' 'us','My','We','Ours','Us','And']
for article in articles:
  ans=0
  for word in article:
    if word in personal_noun:
      ans+=1
  personal_nouns.append(ans)
```

```python
Output_data_structure['FOG INDEX'] = 0.4 * (Output_data_structure['AVG SENTENCE LENGTH'] + Output_data_structure['PERCENTAGE OF COMPLEX WORDS'])
```

```python
Output_data_structure['WORD COUNT'] = words
```

```python
Output_data_structure['SYLLABLE PER WORD'] = np.array(sylabble_counts)/np.array(words)
```

```python
Output_data_structure['COMPLEX WORD COUNT'] = complex_words
```

```python
Output_data_structure['PERSONAL PRONOUNS'] = personal_nouns
```

```python
Output_data_structure['AVG WORD LENGTH'] = np.array(total_characters)/np.array(words)
```
